```python
import numpy as np
import xgboost
import pandas as pd

df = pd.read_csv("./train.csv")
df_valid = pd.read_csv("./validation.csv")
df_test = pd.read_csv("./test.csv")
```

```python
df.head()
```

```python
# train_split = np.array_split(df, 4)
# train = train_split[0]
train = df
train_y = train['click']

cols = ['weekday', 'hour', #'bidid', 'userid', 'useragent',
       #'IP', 
        'region', 'city', 'adexchange', #'domain', 'url', 
        'urlid',
       #'slotid', 
        'slotwidth', 'slotheight', #'slotvisibility', 'slotformat',
       'slotprice', #'creative', 
        'bidprice','payprice', #'keypage',
       'advertiser', 'usertag'
       ]

train_x = train[cols]
```

```python
features = ['weekday', 'hour', 'city', 'slotprice', 'bidprice','payprice', 'advertiser']
#features = cols
X = train[features]
y = train_y

model = xgboost.XGBClassifier()
model.fit(X, y)
```

```python
test_split = np.array_split(df_valid, 1)
test = test_split[0]
test_x = test[features]
test_x
test_predictions = model.predict(test_x)
test_x['click_pred'] = test_predictions
test_x['click'] = test['click']
test_x['abs_diff'] = abs(test_x['click_pred'] - test_x['click'])
test_x[['click_pred', 'click', 'abs_diff']]
```

```python
a=test_x[test_x['click_pred']>0.05]
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test_x['click'], test_x['click_pred'])
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
