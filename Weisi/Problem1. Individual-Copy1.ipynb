```python
def os_browser(datasetToClean):
    print("\nSplit Os and Browser")
    print("Original shape:", datasetToClean.shape)
    os_browser_df=pd.DataFrame(datasetToClean.useragent.str.split('_',1).tolist(), columns=['os','browser'])
    #.str.split('_',1)ï¼Œè¡¨ç¤ºåªåˆ†å‰²ä¸€æ¬¡
    print("Splited as shape:",os_browser_df.shape)

    print("Merging...")
    print("os_browser_df shape:", os_browser_df.shape)
    combineddf=pd.concat([datasetToClean,os_browser_df],axis=1)
    print("Merged as one:",combineddf.shape)

    print("Dropping useragent")
    useragentdropped=combineddf.drop('useragent',axis=1)
    print("Now shape:",useragentdropped.shape)
    
    return useragentdropped
```
