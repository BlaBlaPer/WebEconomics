```python
def Prob1(consp_arr):
        
    Train_Const_summ = pd.DataFrame({'Clicks':'','Impressions':'', 'Spend':'', 'CTR':'', 'CPM':'', 'eCPC':''}, index=[50])
    
    for n in range(len(consp_arr)):
        Train_df['win'] = 0
        Train_df['bidprice'] = consp_arr[n]
        total_budget = 6250000
        budget = 6250000

        for i in Train_df.index:
            current_payprice = Train_df.at[i, 'payprice']
            if  budget>=consp_arr[n]:
                if consp_arr[n]>current_payprice:
                    Train_df.at[i, 'win'] = int(1)
                    budget = budget - current_payprice
            elif budget>0 & budget<consp_arr[n]:
                if budget>current_payprice:
                    Train_df.at[i, 'win'] = int(1)
                    budget = 0
            else:
                #Train_df.loc[i:-1,'bidprice'] = 0
                break
            
        Train_df_new = Train_df[Train_df['win']!=0]
        print(Train_df_new.shape)

        Clk = Train_df_new['click'].sum()
        Imp = Train_df_new['click'].count()
        CTR = Clk/Imp
        Spd = total_budget - budget
        CPM = 1000*Spd/Imp
        eCPC = Spd/Clk
    
        x=20+n
        Train_Const_summ.loc[x,'Clicks']=Clk
        Train_Const_summ.loc[x,'Impressions']=Imp
        Train_Const_summ.loc[x,'Spend']=Spd
        Train_Const_summ.loc[x,'CTR']=CTR
        Train_Const_summ.loc[x,'CPM']=CPM
        Train_Const_summ.loc[x,'eCPC']=eCPC
    

    print(Train_Const_summ)
    
    Train_Const_summ_sort = Train_Const_summ.sort_values(by=['Clicks','CTR','CPM','eCPC'],ascending=[False,False,True,True])
    print(Train_Const_summ_sort)


    return
```
